### Load Price Data

```python
import os
from pathlib import Path
notebook_path = os.getcwd()
current_dir = Path(notebook_path)
csv_file = str(current_dir) + '/VN30F1M_1D.csv'
is_file = os.path.isfile(csv_file)
if is_file:
    dataset = pd.read_csv(csv_file, index_col='Date', parse_dates=True)
else:
    print('remote')
    dataset = pd.read_csv("https://raw.githubusercontent.com/zuongthaotn/vn-stock-data/main/VN30ps/VN30F1M_1D.csv", index_col='Date', parse_dates=True)
```

```python
data = dataset.copy()
```

```python
data
```

```python
data["VolumeK"] = round(data["Volume"] / 1000)
```

```python
data.plot(kind="line", y="VolumeK",  title="Volume line chart by time(K)")
```

```python
data["Money"] = data["Volume"] * data["Close"] * 100000
data["Money_K_Bill"] = round(data["Money"] / 1000000000000)
```
