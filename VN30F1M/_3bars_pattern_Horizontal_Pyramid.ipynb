# Mẫu hình 3 cây nến - Horizontal Pyramid - Analytics

```python
# Horizontal Pyramid Falling
# Điều kiện:

# Nến 1: đỏ (bearish).

# Nến 2: xanh (bullish).

# Nến 3: đỏ (bearish).

# Thân nến: to dần từ nến 1 → nến 2 → nến 3.

# Đặc điểm:

# Đây là một dạng reversal/continuation trap hơn là pattern cổ điển.

# Vì thân nến 2 (xanh) lớn hơn thân nến 1 (đỏ), thể hiện lực hồi phục mạnh. Nhưng sau đó nến 3 đỏ lại lớn hơn cả nến 2, cho thấy phe bán "nuốt trọn" và còn mạnh hơn lực hồi.

# Nó khá giống một biến thể của Bearish Engulfing mở rộng trong 3 nến (Extended Bearish Engulfing) – tức là một cú hồi nhỏ (nến xanh ở giữa) rồi bị lực bán quét sạch bằng cây đỏ mạnh hơn.

# Ý nghĩa giao dịch:

# Xuất hiện ở vùng kháng cự mạnh → xác nhận tín hiệu đảo chiều giảm.

# Xuất hiện trong xu hướng giảm → đóng vai trò continuation (tiếp diễn giảm).

# Thân nến to dần (progressive body) càng củng cố sức mạnh của phe kiểm soát cuối cùng (ở đây là phe bán).
```

```python
data = data[data.index.hour + data.index.minute != 1130]
```

```python
data['color'] = data.apply(lambda r: 'doji' if r["Close"] == r["Open"] else ("green" if r["Close"] > r["Open"] else "red"), axis=1)
```

```python
data['color_s1'] = data['color'].shift(1)
data['color_s2'] = data['color'].shift(2)
```

```python
data['body'] = data.apply(lambda r: abs(r["Close"] - r["Open"]), axis=1)
data['body_s1'] = data['body'].shift(1)
data['body_s2'] = data['body'].shift(2)
```

```python
def get_pattern(r):
    if not (r["body_s2"] < r["body_s1"] and r["body_s1"] < r["body"]):
        return ''
    pattern = ''
    if r['color_s2'] == "red" and r['color_s1'] == "green" and  r['color'] == "red":
        pattern = 'falling'
    elif r['color_s2'] == "green" and r['color_s1'] == "red" and  r['color'] == "green":
        pattern = 'rising'
    return pattern
```

```python
data['pattern'] = data.apply(lambda r: get_pattern(r), axis=1)
```
